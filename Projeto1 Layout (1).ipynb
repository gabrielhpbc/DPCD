```python
base1 = pd.read_csv('PES2011.csv',sep=',')
base5 = pd.read_csv('PES2015.csv',sep=',')
```

```python
base1.V0101=base1.V0101.astype("int")
base5.V0101=base5.V0101.astype("int")
```

```python
base1.loc[(base1.UF<18),"REGIAO"]="NORTE"
base1.loc[(base1.UF>20)&(base1.UF<30),"REGIAO"]="NORDESTE"
base1.loc[(base1.UF>30)&(base1.UF<36),"REGIAO"]="SUDESTE"
base1.loc[(base1.UF>35)&(base1.UF<44),"REGIAO"]="SUL"
base1.loc[(base1.UF>43)&(base1.UF<54),"REGIAO"]="CENTRO-OESTE"
base1.REGIAO=base1.REGIAO.astype("category")
```

```python
base5.loc[(base5.UF<18),"REGIAO"]="NORTE"
base5.loc[(base5.UF>20)&(base5.UF<30),"REGIAO"]="NORDESTE"
base5.loc[(base5.UF>30)&(base5.UF<36),"REGIAO"]="SUDESTE"
base5.loc[(base5.UF>35)&(base5.UF<44),"REGIAO"]="SUL"
base5.loc[(base5.UF>43)&(base5.UF<54),"REGIAO"]="CENTRO-OESTE"
base5.REGIAO=base5.REGIAO.astype("category")
```

```python
base1.loc[(base1.V0601==1) | (base1.V0602==1),'Alfabetização Básica'] = 'Sim'
base1.loc[(base1.V0601==3) | (base1.V0602==3),'Alfabetização Básica'] = 'Parcial'
base1.loc[(base1.V0601==3) & (base1.V0602==3),'Alfabetização Básica'] = 'Não'
base1.V0601=base1.V0601.astype("category")
base1.V0601=base1.V0601.astype("category")
```

```python
base5.loc[(base5.V0601==1) | (base5.V0602==1),'Alfabetização Básica'] = 'Sim'
base5.loc[(base5.V0601==3) | (base5.V0602==3),'Alfabetização Básica'] = 'Parcial'
base5.loc[(base5.V0601==3) & (base5.V0602==3),'Alfabetização Básica'] = 'Não'
base5.V0601=base5.V0601.astype("category")
base5.V0601=base5.V0601.astype("category")
```
